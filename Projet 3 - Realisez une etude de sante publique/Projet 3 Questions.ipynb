```python
#Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal.

#Question 3

data = pd.read_csv("C:/Users/KyRun69/Downloads/population.csv")
df=pd.DataFrame(data)

pop_2013=df[df["Année"]==2013]
tableau={"Pays":pop_2013.Zone.tolist(),"Population":pop_2013.Valeur.tolist()}
pop_2013=pd.DataFrame(tableau)
pop_2013.Population=pop_2013.Population*1000

df_bilan = pd.read_csv("C:/Users/KyRun69/Downloads/Bilan_alimentaire.csv")


df_dispo=df_bilan[df_bilan["Code Élément"]==664]
df_dispo=pd.merge(df_dispo,pop_2013,how="left")
df_dispo["Disponibilité alimentaire (Kcal)"]=(df_dispo.Valeur)*(df_dispo.Population)*365


print(df_dispo)
```

```python
#Question 4  calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg.

df_transition=df_bilan[df_bilan["Code Élément"]==645]
df_transition=df_transition.reset_index(drop=True)
valeur_transition={"Pays":df_transition.Pays.tolist(),"Code Produit":df_transition["Code Produit"].tolist(),"Disponibilité alimentaire en quantité":df_transition.Valeur.tolist()}
df_transition=pd.DataFrame(valeur_transition)


df_dispo=pd.merge(df_dispo,df_transition,how="left")
df_dispo=df_dispo.replace(0,np.nan)
df_dispo=df_dispo.dropna(axis=0, how='any')
df_dispo["Ratio"]=(df_dispo["Disponibilité alimentaire (Kcal)"])/((df_dispo["Disponibilité alimentaire en quantité"])*(df_dispo.Population))
ratio={"Code produit":df_dispo["Code Produit"].tolist(),"Produit":df_dispo.Produit.tolist(),"Pays":df_dispo.Pays.tolist(),"Ratio":df_dispo.Ratio.tolist()}
df_ratio=pd.DataFrame(ratio)
df_ratio=df_ratio.replace(0,np.nan)
df_ratio=df_ratio.dropna(axis=0, how='any')
df_ratio=df_ratio.reset_index(drop=True)

Table_ratio=df_ratio.groupby("Produit").mean()
Table_ratio=Table_ratio.reset_index()

print(Table_ratio)

```

```python
# Question 6  Calculez, pour les produits végétaux uniquement, la disponibilité intérieure mondiale exprimée en kcal.

data_2 = pd.read_csv("C:/Users/KyRun69/Downloads/bilan_alimentaire_vegetaux.csv")
df_vegetaux=pd.DataFrame(data_2)
df_dispo2=df_vegetaux[df_vegetaux["Code Élément"]==5301]
df_dispo2=df_dispo2.replace(0,np.nan)
df_dispo2=df_dispo2.dropna(axis=0, how='any')
df_dispo2=df_dispo2.reset_index(drop=True)
disponibilite={"Produit":df_dispo2.Produit.tolist(),"Valeur_kg":df_dispo2.Valeur.tolist()}
df_disponibilite=pd.DataFrame(disponibilite)
df_disponibilite.Valeur_kg=df_disponibilite.Valeur_kg*1000000
df_disponibilite=df_disponibilite.groupby("Produit").Valeur_kg.sum()
df_disponibilite=df_disponibilite.reset_index()
df_disponibilite=pd.merge(df_disponibilite,Table_ratio,how="left")
df_disponibilite["Valeur_Kcal"]=df_disponibilite.Valeur_kg*df_disponibilite.Ratio
df_disponibilite=df_disponibilite.replace([np.inf], np.nan)
df_disponibilite=df_disponibilite.dropna(axis=0,how="any")

somme_dispo_vegetaux=df_disponibilite["Valeur_Kcal"].sum()

print("La disponibilité intérieure mondiale des produits végétaux est de "+ str(somme_dispo_vegetaux) + " Kcal")
```

```python
#Question 8   En termes de disponibilité intérieure de produits végétaux, combien d'humains pourraient être nourris si la production végétale, mis à part ses utilisations en Semence, Traitement et Autres utilisation, était utilisée pour de la nourriture ?


df_vegetaux_sauf_semences=df_vegetaux[df_vegetaux["Code Élément"]==5527]
df_vegetaux_sauf_semences.Valeur=df_vegetaux_sauf_semences.Valeur*1000000
df_vegetaux_sauf_semences=df_vegetaux_sauf_semences.groupby("Produit").Valeur.sum()
df_vegetaux_sauf_semences=df_vegetaux_sauf_semences.reset_index()
df_vegetaux_sauf_semences=pd.merge(df_vegetaux_sauf_semences,Table_ratio,how="left")
df_vegetaux_sauf_semences["Kcal"]=df_vegetaux_sauf_semences.Valeur*df_vegetaux_sauf_semences.Ratio
df_vegetaux_sauf_semences=df_vegetaux_sauf_semences.replace([np.inf], np.nan)
df_vegetaux_sauf_semences=df_vegetaux_sauf_semences.dropna(axis=0,how="any")


df_vegetaux_sauf_autre=df_vegetaux[df_vegetaux["Code Élément"]==5154]
df_vegetaux_sauf_autre.Valeur=df_vegetaux_sauf_autre.Valeur*1000000
df_vegetaux_sauf_autre=df_vegetaux_sauf_autre.groupby("Produit").Valeur.sum()
df_vegetaux_sauf_autre=df_vegetaux_sauf_autre.reset_index()
df_vegetaux_sauf_autre=pd.merge(df_vegetaux_sauf_autre,Table_ratio,how="left")
df_vegetaux_sauf_autre["Kcal"]=df_vegetaux_sauf_autre.Valeur*df_vegetaux_sauf_autre.Ratio
df_vegetaux_sauf_autre=df_vegetaux_sauf_autre.replace([np.inf], np.nan)
df_vegetaux_sauf_autre=df_vegetaux_sauf_autre.dropna(axis=0,how="any")


df_vegetaux_sauf_traitement=df_vegetaux[df_vegetaux["Code Élément"]==5131]
df_vegetaux_sauf_traitement.Valeur=df_vegetaux_sauf_traitement.Valeur*1000000
df_vegetaux_sauf_traitement=df_vegetaux_sauf_traitement.groupby("Produit").Valeur.sum()
df_vegetaux_sauf_traitement=df_vegetaux_sauf_traitement.reset_index()
df_vegetaux_sauf_traitement=pd.merge(df_vegetaux_sauf_traitement,Table_ratio,how="left")
df_vegetaux_sauf_traitement["Kcal"]=df_vegetaux_sauf_traitement.Valeur*df_vegetaux_sauf_traitement.Ratio
df_vegetaux_sauf_traitement=df_vegetaux_sauf_traitement.replace([np.inf], np.nan)
df_vegetaux_sauf_traitement=df_vegetaux_sauf_traitement.dropna(axis=0,how="any")

valeur_dispo_vegetaux=somme_dispo_vegetaux-(df_vegetaux_sauf_semences.Kcal.sum()+df_vegetaux_sauf_autre.Kcal.sum()+df_vegetaux_sauf_traitement.Kcal.sum())
print("La disponibilité intérieure mondiale des produits végétaux en omettant les catégories semences, traitement et autre est de "+ str(valeur_dispo_vegetaux) + " Kcal")
print("Il serait alors possible de nourrire " + str(int(valeur_dispo_vegetaux/besoin_pop_moyen)) + " humains") 

```

```python
#Question 11 :De la même manière que vous avez calculé la table des ratios énergie/poids , calculez les ratios poids de protéines/poids total, puis convertissez ces données (filtrées) en terme de protéines (unité : gramme).

prot=pd.read_csv("C:/Users/KyRun69/Downloads/prot_monde.csv")
proteines=pd.DataFrame(prot)
proteines=pd.merge(proteines,pop_2013,how="left")
print(proteines)
proteines["Valeur"]=(proteines.Valeur*proteines.Population)
proteines=proteines.drop(columns="Population")
proteines=proteines.groupby("Produit").Valeur.sum()
proteines=proteines.reset_index()
proteines=proteines.rename(index=str, columns={"Valeur": "Protéines"})


dispoo=pd.read_csv("C:/Users/KyRun69/Downloads/question_11.csv")
df_dispoo=pd.DataFrame(dispoo)
df_dispoo=pd.merge(df_dispoo,pop_2013,how="left")
df_dispoo["Valeur"]=df_dispoo.Valeur*df_dispoo.Population/365
df_dispoo=df_dispoo.groupby("Produit").Valeur.sum()
df_dispoo=df_dispoo.reset_index()
df_dispoo=df_dispoo.rename(columns={"Valeur":"Disponibilité alimentaire en quantité"})



proteines=pd.merge(proteines,df_dispoo,how="left")
proteines["Ratio Proteines/Poids"]=proteines["Protéines"]/proteines["Disponibilité alimentaire en quantité"]
proteines=proteines.dropna(axis=0,how="any")

print(proteines)
```
